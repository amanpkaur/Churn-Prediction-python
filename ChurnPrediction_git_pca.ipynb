## Understanding the Datasets

### Train vs. Test
In this competition, you’ll gain access to two datasets that are samples of past subscriptions of a video streaming platform that contain information about the customer, the customers streaming preferences, and their activity in the subscription thus far. One dataset is titled `train.csv` and the other is titled `test.csv`.

Both `train.csv` and `test.csv` contain one row for each unique subscription. For each subscription, a single observation (`CustomerID`) is included during which the subscription was active. 

In addition to this identifier column, the `train.csv` dataset also contains the label label for the task, a binary column `Churn`.

#### We will add Principal Component Analysis to the numerical columns before applying any of the models.
First, let us select the number of components needed to apply the PCA for each model.

```python
# Apply PCA to the numerical columns to calculate the number of components.
train_num_scaled=train_num_df_scaled.drop(columns="type")
test_num_scaled=test_num_df_scaled.drop(columns="type")
pca = PCA().fit(train_num_scaled)
x_num = np.arange(1, len(train_num_scaled.columns)+1, step=1)
y_num = np.cumsum(pca.explained_variance_ratio_)
desired_x=np.interp(0.9, y_num, x_num)

# Apply PCA to categorical hot encoded columns
train_cat_scaled=train_df_noid_dummy_scaled[train_df_noid_dummy_scaled.columns.difference(num_df.columns)]
test_cat_scaled=test_df_noid_dummy_scaled[test_df_noid_dummy_scaled.columns.difference(num_df.columns)]
pca_cat = PCA().fit(train_cat_scaled)
x_cat = np.arange(1, len(train_cat_scaled.columns)+1, step=1)
y_cat = np.cumsum(pca_cat.explained_variance_ratio_)

desired_x_cat=np.interp(0.9, y_cat, x_cat)


# Create subplots for both the categorical and numerical data
fig, axes = plt.subplots(1,2, figsize=(12, 6))

# Plot for numerical columns
ax1 = axes[0]
ax1.plot(x_num, y_num, marker='o', linestyle='--', color='b')
ax1.set_xlabel('Number of Components')
ax1.set_ylabel('Cumulative variance (%)')
ax1.set_title('Numerical Columns - Explained Variance')
ax1.axhline(y=0.95, color='r', linestyle='-', label='95% cut-off threshold')
ax1.text(0.65, 0.85, '95% cut-off threshold', color='red', fontsize=12)
ax1.legend()
ax1.grid(axis="y")

# Plot for categorical columns
ax2 = axes[1]
ax2.plot(x_cat, y_cat, marker='o', linestyle=':', color='b')
ax2.set_xlabel('Number of Components')
ax2.set_ylabel('Cumulative variance (%)')
ax2.set_title('Categorical Columns - Explained Variance')
ax2.axhline(y=0.95, color='r', linestyle='-', label='95% cut-off threshold')
ax2.text(0.55, 0.85, '95% cut-off threshold', color='red', fontsize=12)
ax2.legend()
ax2.grid(axis="y")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# we will apply the pca transformation to both the numerical and the categorical columns
def apply_pca(data_scaled, n_components, column_prefix,test_data_scaled):
    pca = PCA(n_components=n_components)
    data_pca = pca.fit_transform(data_scaled)
    pca_columns = [f'{column_prefix}{i + 1}' for i in range(n_components)]
    features_pca = pd.DataFrame(data_pca, columns=pca_columns)
    explained_variance = pca.explained_variance_ratio_
    loadings = pca.components_
    features_test_data_pca = pd.DataFrame(pca.transform(test_data_scaled), columns=pca_columns)
    return features_pca, explained_variance,  features_test_data_pca,loadings

# Apply PCA to numerical columns
n_components_num = math.ceil(desired_x)
features_train_num_pca, explained_variance_num,features_test_num_pca,loadings_num = apply_pca(train_num_scaled, n_components_num, 'PC_num',test_num_scaled)
# Apply PCA to categorical columns
n_components_cat = math.ceil(desired_x_cat)
features_train_cat_pca, explained_variance_cat,features_test_cat_pca,loadings_cat = apply_pca(train_cat_scaled, n_components_cat, 'PC_cat',test_cat_scaled)
```

```python
#PCA features for training
features_train_pca=pd.concat([features_train_num_pca,features_train_cat_pca],axis=1)
features_test_pca=pd.concat([features_test_num_pca,features_test_cat_pca],axis=1)
```

#### Are the data imbalanced?

```python
train_df_noid_dummy_scaled.Churn.value_counts()

non_churned = (
    train_df_noid_dummy_scaled.Churn.value_counts()[0]
    / train_df_noid_dummy_scaled.Churn.value_counts().sum()
)
churned = 1 - non_churned
sns.countplot(train_df_noid_dummy_scaled, x="Churn")
plt.title(f"Labeled Data Count (Ratio={non_churned:.2f} : {churned:.2f})")
plt.legend()
plt.tight_layout()
```

As seen above, the data are highly imbalanced. Let us apply SMOTE to balance the dataset.

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
features = features_train_pca
labels = train_df_noid_dummy_scaled.Churn
features_smoted, labels_smoted = sm.fit_resample(features, labels)
# check if SMOTE balanced the two classes
labels_smoted.value_counts()[0] = labels_smoted.value_counts()[1]
```

```python
# Split the data
features_train, features_val, labels_train, labels_val = train_test_split(
    features_smoted, labels_smoted, test_size=0.2
)
```

We will use the `RandomForestClassifier` from the `sklearn` library employing 1000 decision trees with leaf splitting to a depth of 2. 

```python
model_rf = RandomForestClassifier(
    n_estimators=50, criterion="gini", random_state=42, max_depth=2, min_samples_leaf=2
)
model_rf.fit(features_train, labels_train)
```
